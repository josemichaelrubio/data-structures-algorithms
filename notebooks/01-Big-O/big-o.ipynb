# Big-O

**Big-O Asymptotic analysis**

Big-O tells us how well a problem is solved

It's the notation we use to see how long an algorithm runs 



Thus, since every machine is different. We measure how many operations the code runs instead of the time it takes to complete the algorithm

For Big-O, we don't care how steep the line is, we care about about the line moves

## Rule Book for calc Big-O

1. **Worst Case**
    - When calculating Big-O, we always think about the worst case. 
2. **Remove Constants**
    - Constants don't matter in the grand scheme of things, because if we loop through an infinate amount of those times, those constants are insignifcant O(n) + n 
3. **Different Terms for inputs**
    - It depends on the differnt inputs that are being looped O(a+b)
4. Drop Non-Dominants
